```python
def generate_themis_plot_links(event_df, num_links, satellite, event_class_label, random_state=None):
    df_filtered = event_df.query(f"event_class == {event_class_label} and event_duration < 120").copy()
    
    df_sample = df_filtered.sample(num_links, random_state=random_state).sort_values(by='start_time', ascending=True).copy()

    for _, row in df_sample.iterrows():
        start_time = pd.to_datetime(row['start_time'])
        end_time = pd.to_datetime(row['end_time'])

        year = start_time.year
        month = f"{start_time.month:02d}"
        day = f"{start_time.day:02d}"

        print(f"{str(start_time)[:19]} - {str(end_time.time())[:8]}")
        print(f"https://themis.ssl.berkeley.edu/summary.php?year={year}&month={month}&day={day}&hour=0024&sumType={satellite}&type=overview\n")

generate_themis_plot_links(event_df=event_df, num_links=0, satellite='tha', event_class_label=3)
```
